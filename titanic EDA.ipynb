I wanted to see if there was any preverances made based on the calss in which passenger travelled.

```python
# Comparing the survival rate based on class 

df[(df.Pclass==1)].Survived.value_counts().plot(lw=2,kind="barh",title="Passengers Survived",facecolor='grey', alpha=0.5)

df[(df.Pclass==2)].Survived.value_counts().plot(lw=2,kind="barh",facecolor='red', alpha=0.3)

df[(df.Pclass==3)].Survived.value_counts().plot(lw=2,kind="barh",facecolor='yellow', alpha=0.1)

plt.legend(["Pclass 1", "Pclass 2", "Pclass 3"])
```
